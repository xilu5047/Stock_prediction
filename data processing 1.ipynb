```python
import yfinance as yf
import pandas as pd
import pandas_datareader.data as web
from datetime import datetime
from pandas_datareader.oecd import OECDReader

start_date = datetime(2020, 1, 1)
end_date = datetime(2022, 12, 31)

companies = ["IBM","AAPL","JPM"]

all_data = pd.DataFrame()

all_financials = pd.DataFrame()

for company in companies:
    data = yf.download(company, start=start_date, end=end_date)
    data['Company'] = company
    all_data = pd.concat([all_data, data])
    
    ticker = yf.Ticker(company)
    financials = ticker.quarterly_financials.T 
    financials['Company'] = company
    financials['Date'] = financials.index
    all_financials = pd.concat([all_financials, financials])
```

```python
import requests
import pandas as pd

url = 'https://www.alphavantage.co/query?function=OVERVIEW&symbol=IBM&apikey=LT44MKGZSR0QGXKV'
r = requests.get(url)
data = r.json()

numerical_features = ['MarketCapitalization', 'EBITDA', 'PERatio', 'PEGRatio', 'BookValue', 'DividendPerShare',
                      'DividendYield', 'EPS', 'RevenuePerShareTTM', 'ProfitMargin', 'OperatingMarginTTM',
                      'ReturnOnAssetsTTM', 'ReturnOnEquityTTM', 'RevenueTTM', 'GrossProfitTTM', 'DilutedEPSTTM',
                      'QuarterlyEarningsGrowthYOY', 'QuarterlyRevenueGrowthYOY', 'TrailingPE', 'ForwardPE',
                      'PriceToSalesRatioTTM', 'PriceToBookRatio', 'EVToRevenue', 'EVToEBITDA', 'Beta', 'SharesOutstanding']

numerical_data = {key: float(data[key]) for key in numerical_features}
numerical_df = pd.DataFrame([numerical_data])

print(numerical_df)
```

```python
import requests
import pandas as pd

def fetch_company_data(url):
    response = requests.get(url)
    data = response.json()
    return data

def process_company_data(data):
    numerical_features = ['MarketCapitalization', 'EBITDA', 'PERatio', 'PEGRatio', 'BookValue', 'DividendPerShare',
                          'DividendYield', 'EPS', 'RevenuePerShareTTM', 'ProfitMargin', 'OperatingMarginTTM',
                          'ReturnOnAssetsTTM', 'ReturnOnEquityTTM', 'RevenueTTM', 'GrossProfitTTM', 'DilutedEPSTTM',
                          'QuarterlyEarningsGrowthYOY', 'QuarterlyRevenueGrowthYOY', 'TrailingPE', 'ForwardPE',
                          'PriceToSalesRatioTTM', 'PriceToBookRatio', 'EVToRevenue', 'EVToEBITDA', 'Beta', 'SharesOutstanding']
    
    categorical_features = ['Sector', 'Industry', 'Country', 'Symbol']
    
    numerical_data = {}
    for key in numerical_features:
        try:
            numerical_data[key] = float(data.get(key, 0))
        except (ValueError, TypeError):
            numerical_data[key] = 0.0  
    
    categorical_data = {key: data.get(key, '') for key in categorical_features}
    
    return {**numerical_data, **categorical_data}

def process_multiple_urls(urls):
    all_data = []
    for url in urls:
        data = fetch_company_data(url)
        processed_data = process_company_data(data)
        all_data.append(processed_data)
    
    df = pd.DataFrame(all_data)
    
    return df

urls = [
    'https://www.alphavantage.co/query?function=OVERVIEW&symbol=IBM&apikey=PYOH5XKNWMHBM4A3',  # Technology
    'https://www.alphavantage.co/query?function=OVERVIEW&symbol=AAPL&apikey=PYOH5XKNWMHBM4A3', # Technology
    'https://www.alphavantage.co/query?function=OVERVIEW&symbol=JPM&apikey=PYOH5XKNWMHBM4A3', # Financial
    'https://www.alphavantage.co/query?function=OVERVIEW&symbol=BA&apikey=PYOH5XKNWMHBM4A3'  # Aerospace
]

# 处理多个URL
df_final = process_multiple_urls(urls)
print(df_final)
```
